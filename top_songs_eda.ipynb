```python
# Loading the dependencies
import pandas as pd
```

```python
# Reading the csv file as a DataFrame
df = pd.read_csv('Resources/Top 500 Songs.csv')
df
```

```python
# Looking at the data types
df.dtypes
```

```python
# Looking at the values in released
df['released'].value_counts()
```

```python
# Splitting released into two columns
# Correcting my syntax with extra set of brackets, https://stackoverflow.com/questions/37333299/splitting-a-pandas-dataframe-column-by-delimiter
df[['released_month', 'released year']] = df['released'].str.split(',', expand=True)
df
```

```python
# Dropping original released column
df = df.drop(columns='released')
df
```

```python
# Looking at values in released_year
df['released year'].value_counts()
```

```python
# Looking at values in released_month column
df['released_month'].value_counts()
```

```python
# Dropping the month column
df = df.drop(columns='released_month')
```

```python
# Looking at values in position column
df['position'].value_counts()
```

```python
# Splitting position column into two columns
df[['.', 'peak_position']] = df['position'].str.split(expand=True)
df
```

```python
# Looking at the values in peak_position
df['peak_position'].value_counts()
```

```python
# Dropping the unneeded columns
df = df.drop(columns=['position', '.'])
df
```

```python
# Looking at the values in streak
df['streak'].value_counts()
```

```python
df['streak'] = df['streak'].str.replace(' weeks', '')
df
```

```python
# Looking at values in streak column
df['streak'].value_counts()
```

```python
# Dropping streak column for this analysis
df = df.drop(columns='streak')
df
```

```python
# Exporting clean data as a csv file
df.to_csv('Resources/clean_top_songs.csv', index=False)
```
